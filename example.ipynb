## Randomness in the generated text

```python
genai.configure(api_key=os.getenv('GOOGLE_API_KEY'))
model1=genai.GenerativeModel('gemini-2.5-flash-lite')
print(model1.generate_content('Tell me the joke on the affair between PTE sir and ENglish MAam').text)
```

```python
genai.configure(api_key=os.getenv('GOOGLE_API_KEY'))
model1=genai.GenerativeModel('gemini-2.5-flash-lite')
prompt='Tell me the bad joke on engineering'
response=model1.generate_content(prompt,generation_config={'top_p':1})
print(response.text)
```

# Methods to control randomness in the output

1)Temperature (Controls the randomness and creativity)
* 0 means deterministic and focused 
* 1 means it gives balanced creativity
* greater than 1 is more diversed and less coherent 

2)top_p(nucleus sampling - probability cut off)

3)top_k(sample from top k likely tokens)
